```python
from robomimic.config import config_factory
from robomimic.scripts.train import train
import robomimic.utils.torch_utils as TorchUtils
import warnings
warnings.filterwarnings("ignore")

# start as BC
config = config_factory(algo_name="bc")
config.unlock()

config.train.hdf5_cache_mode = "low_dim" 
# config.train.hdf5_cache_mode = None 
config.train.num_data_workers = 0 

# ===== DATASET CONFIG =====
config.train.data = "datasets/resident_evil_dataset.hdf5"
config.train.output_dir = "logs/resident_evil/bc_transformer"
config.train.batch_size = 32 # 64 
config.train.num_epochs = 300
config.train.seed = 42
config.train.seq_length = 1
config.train.frame_stack = 24 
config.train.pad_frame_stack = True
config.train.hdf5_normalize_actions = False


config.train.dataset_keys = ["actions", "rewards", "dones"]
config.all_obs_keys = ["image"]

config.observation.modalities.obs.rgb = ["image"]
config.observation.modalities.obs.low_dim = []


# ===== ALGO CONFIG (TRANSFORMER) =====
config.algo.optim_params.policy.optimizer_type = "adamw"
config.algo.optim_params.policy.learning_rate.initial = 1e-4

# Transformer
config.algo.transformer.enabled = True
config.algo.transformer.context_length = 24
config.algo.transformer.embed_dim = 512
config.algo.transformer.num_layers = 6
config.algo.transformer.num_heads = 8
config.algo.transformer.emb_dropout = 0.1
config.algo.transformer.attn_dropout = 0.1
config.algo.transformer.block_dropout = 0.1

# GMM/Gaussian costuma ser instável para botões, vamos desativar
config.algo.gaussian.enabled = False

config.experiment.rollout.enabled = False
config.experiment.env = None
config.train.hdf5_load_next_obs = False

# ===== VISION ENCODER (ResNet18) =====
config.observation.modalities.obs.rgb = ["image"]
config.observation.encoder.rgb.core_class = "VisualCore"
config.observation.encoder.rgb.core_kwargs = {
    "feature_dimensions": 512,
    "backbone_class": "ResNet18Conv",
    "backbone_kwargs": {
        "pretrained": True,
        "input_coord_conv": True
    },
    "pool": False
}


config.algo.horizon.observation_horizon = 24
config.algo.horizon.prediction_horizon = 1

config.lock()

# Train
device = TorchUtils.get_torch_device(try_to_use_cuda=True)
print(f"🚀 Iniciando Treino BC-Transformer na {device}...")
train(config, device=device)
```
